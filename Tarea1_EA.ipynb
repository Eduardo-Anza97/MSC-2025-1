```python
#4. En una nueva celda: Pedirle al usuario ingresar una lista de números reales positivos que correspondan a valores de el radio de una circunferencia.
#5. Utilice except, para repetir la pregunta si no es un numero real positivo
#6. Calcule el área y perímetro de ellos
#7. Imprima en pantalla un mensaje indicando el area y el perimétro, el lenguaje debe ser claro.

"""
Sin ciclo For
"""

print("Hola, bienvenido/a.")
print("")

pi = 3.141592653

while True:
        try:
            radio = float(input("Ingrese el radio de la circunferencia, recuerde debe ser un número real positivo: ")) #4.
            assert radio > 0
            if radio > 0:
              Area = pi*(radio**2) #6.
              Perimetro = 2*pi*radio #6.
              print("La Área es: ",round(Area,2)," unidades cuadradas") #7.
              print("El Perimetro es: ",round(Perimetro,2)," unidades") #7.
              print("")
            break
        except AssertionError: #5.
          print("¡Error! : Ingrese el radio de una circunferencia que sea un número real positivo")
          print("")
        except ValueError:
          print("¡Error! : Ingrese el radio de una circunferencia que sea un número real positivo")
          print("")
```
