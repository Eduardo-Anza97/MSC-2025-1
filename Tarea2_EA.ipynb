```python
#Grafico 2D
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10,10,100)

plt.plot(x, x**2, label='$ f(x) = x^2$')
plt.title('''Grafica de $f(x)=x^2$''')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.legend(loc=4)
plt.savefig('graficos/grafica2D.png') #Correción: Se guarda en carperta creada a través de codigo
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Crear una malla de puntos en los ejes X y Y
x = np.linspace(-5,5,100) # Valores de x
y = np.linspace(-5,5,100) # Valores de y
x, y = np.meshgrid(x, y) # Crear la malla de puntos 2D
z = np.sin(x)*np.cos(y) # Definir la funcion

# Crear el gráfico 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Graficar la superficie
ax.plot_surface(x, y, z, cmap='PuRd')

# Configurar etiquetas de los ejes
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')
ax.set_title('Grafica de $f(x,y)=sin(x)cos(y)$')

# Mostrar el gráfico
plt.savefig('graficos/grafica3D.png') #Correción: Se guarda en carperta creada a través de codigo
plt.show()
```
